```python
import pandas as pd
import numpy as np
import glob
```

```python
merge_csv_path='/content/drive/MyDrive/Project_AIL303_GROUP2/ml-32m/merged_checkpoints.csv'
file_checkpoint_list = glob.glob('/content/drive/MyDrive/Project_AIL303_GROUP2/checkpoints_poster_csv/*.csv')
df_list = [ pd.read_csv(file_) for file_ in file_checkpoint_list]
merged_df = pd.concat(df_list, ignore_index=True)
l0= len(merged_df)
result_checkpoint_df = merged_df.drop_duplicates(subset=['imdb_url'], keep= 'first')
result_checkpoint_df.dropna(inplace=True)
unwanted_values = ['', 'Timeout when waiting for JSON']
result_checkpoint_df = result_checkpoint_df[~result_checkpoint_df['poster_link'].isin(unwanted_values)]
print(f'loai bo trung lap:{l0-len(result_checkpoint_df)}')
result_checkpoint_df.to_csv(merge_csv_path, index=False)
```

```python
# Code de add vao dataset chinh
dt= pd.read_csv('/content/drive/MyDrive/Project_DAP391m_Group2/Dataset/ml-32m/merged_dataset(2).csv')
dt.isnull().sum()
print(dt.head())
```

```python
df_checkpoint = pd.read_csv('/content/drive/MyDrive/Project_AIL303_GROUP2/checkpoints_poster_csv/merged_checkpoint.csv')
all_poster_links = pd.Series(df_checkpoint.poster_link.values, index=df_checkpoint.imdb_url).to_dict()
dt['poster_link'] = dt['imdb'].map(all_poster_links)
dt.isnull().sum()
```

```python
print(len(dt['movieId'].unique()))
print(len(dt['poster_link'].unique()))
```

```python
def count_dup_column(dt_cleaned):
  for ten_cot in dt_cleaned.columns:
      so_trung_lap = dt_cleaned.duplicated(subset=[ten_cot]).sum()
      print(f"- Cột '{ten_cot}': có {so_trung_lap} giá trị trùng lặp.")
```

```python
dt.drop('imdbId', axis = 1, inplace= True)
dt.dropna(inplace=True)
dt_cleaned = dt[dt['poster_link']!= '']
count_dup_column(dt_cleaned)
dt_cleaned.duplicated().sum()
```

```python
thu_tu_new_columns=['userId', 'movieId', 'rating', 'title', 'poster_link']
dt_cleaned = dt_cleaned[thu_tu_new_columns]
```
